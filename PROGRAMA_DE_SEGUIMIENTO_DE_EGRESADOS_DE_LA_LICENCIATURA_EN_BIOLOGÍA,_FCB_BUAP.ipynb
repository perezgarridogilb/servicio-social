```
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import io
import glob 
from google.colab import drive
num_format = "{:.2f}".format
drive.mount('/content/gdrive')
egresados = pd.read_csv('gdrive/My Drive/csv/INSCRIPCIÓN EGRESADOS. _DÍA DEL BIÓLOGO 2021_ (respuestas).csv', header = 0)
estudiantes = pd.read_csv('gdrive/My Drive/csv/INSCRIPCIÓN ESTUDIANTES. _DÍA DEL BIÓLOGO 2021_ (respuestas).csv', header = 0)
```

```
#@title Egresados
egresados
```

```
#@title Estudiantes
estudiantes
```

```
#@title 1.- Analizar en egresados y estudiantes la edad promedio de los asistentes.
print('Edad promedio de los egresados:', num_format(egresados['Edad'].mean()))
print('Edad promedio de los estudiantes:', num_format(estudiantes['Edad, años'].mean()))
```

```
#@title 2.- Analizar en cada caso, cuantos hombres y cuantas mujeres asistieron (también en %)
hombresEgresados = egresados[egresados['Sexo'] == 'Hombre']
mujeresEgresadas = egresados[egresados['Sexo'] == 'Mujer']
hombresEstudiantes = estudiantes[estudiantes['Sexo'] == 'Hombre']
mujeresEstudiantas = estudiantes[estudiantes['Sexo'] == 'Mujer']
print(f"Porcentaje de hombres egresados: {num_format(float((len(hombresEgresados) / len(egresados)) * 100))}%   Porcentaje de mujeres egresadas: {num_format(float((len(mujeresEgresadas) / len(egresados)) * 100))}%")
print(f"Porcentaje de hombres estudiantes: {num_format(float((len(hombresEstudiantes) / len(estudiantes)) * 100))}% Porcentaje de mujeres estudiantas: {num_format(float((len(mujeresEstudiantas) / len(estudiantes)) * 100))}%")
```

```
#@title 3.- Analizar en cada caso, la edad de hombres y mujeres por separado (también en %)
countEdadEgresadosHombres = hombresEgresados['Edad'].value_counts()
countEdadEgresadasMujeres = mujeresEgresadas['Edad'].value_counts()
countEdadEstudiantesHombres = hombresEstudiantes['Edad, años'].value_counts()
countEdadEstudiantasMujeres = mujeresEstudiantas['Edad, años'].value_counts()
print(f"\t\t  Egresados \n\nPara {len(hombresEgresados)} hombres:")
for x in range(len(countEdadEgresadosHombres)):
    print(f"La edad de {int(countEdadEgresadosHombres.index[x])} años tiene un porcentaje del: {num_format((countEdadEgresadosHombres.iloc[x] / len(hombresEgresados)) * 100)}%")
print(f"\nPara {len(mujeresEgresadas)} mujeres:")
for x in range(len(countEdadEgresadasMujeres)):
    print(f"La edad de {int(countEdadEgresadasMujeres.index[x])} años tiene un porcentaje del: {num_format((countEdadEgresadasMujeres.iloc[x] / len(mujeresEgresadas)) * 100)}%")
print(f"\n\t\t  Estudiantes \n\nPara {len(hombresEstudiantes)} hombres:")
for x in range(len(countEdadEstudiantesHombres)):
    print(f"La edad de {int(countEdadEstudiantesHombres.index[x])} años tiene un porcentaje del: {num_format((countEdadEstudiantesHombres.iloc[x] / len(hombresEstudiantes)) * 100)}%")
print(f"\nPara {len(mujeresEstudiantas)} mujeres:")
for x in range(len(countEdadEstudiantasMujeres)):
    print(f"La edad de {int(countEdadEstudiantasMujeres.index[x])} años tiene un porcentaje del: {num_format((countEdadEstudiantasMujeres.iloc[x] / len(mujeresEstudiantas)) * 100)}%")
```

```
#@title 5.- En egresados, en otra carpeta ordenar (filtrar la información por generación de ingreso), para cuantificar cuantas personas son de cada generación. Ingresar un gráfico de barras. Esto nos permitirá observar desde que generación tenemos presencia. Tomar en cuenta que la primera generación es de 1987

generaciones = egresados['Generación de ingreso'].value_counts()
porGeneracion = generaciones.sort_index().index
porCantidad = generaciones.sort_index().iloc
#print('generaciones.sort_index()')
fig, ax = plt.subplots(figsize=(14,9))
ax.set_title('Cantidad de egresados por generaciones')
ax.set_xticks(porGeneracion)
plt.bar(porGeneracion, porCantidad, align = 'center', width =0.7)
plt.xticks(rotation= 30)
plt.yticks(np.arange(0, 16, 1))
plt.savefig('barras.png')
plt.show()
```
